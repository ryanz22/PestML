```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
actual_values = ['dog', 'cat', 'dog', 'cat', 'dog', 'dog', 'cat', 'dog']
predicted_values = ['dog', 'dog', 'dog', 'cat', 'dog', 'dog', 'cat', 'cat']

cm = confusion_matrix(actual_values, predicted_values)
print(cm)
```

```python
print(classification_report(actual_values, predicted_values))
```

```python
y =      [1, 2, 2, 3, 2, 1, 3]
pred_y = [1, 3, 2, 2, 3, 1, 1]

cm2 = confusion_matrix(y, pred_y)
print(cm2)

disp = ConfusionMatrixDisplay(confusion_matrix=cm2,
                              display_labels=['dog', 'cat', 'pig'])
disp.plot()
plt.show()
```

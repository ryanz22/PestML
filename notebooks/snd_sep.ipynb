# Source Separation

```python
from speechbrain.pretrained import SepformerSeparation as separator
import speechbrain as sb
from speechbrain.dataio.dataio import read_audio
from IPython.display import Audio
import IPython
```

```python

model = separator.from_hparams(source="speechbrain/sepformer-wsj02mix", savedir='pretrained_models/sepformer-wsj02mix')
est_sources = model.separate_file(path='speechbrain/sepformer-wsj02mix/test_mixture.wav')
```

```python
signal = read_audio("test_mixture.wav").squeeze()
Audio(signal, rate=8000)
```

```python
Audio(est_sources[:, :, 0].detach().cpu().squeeze(), rate=8000)
```

```python
Audio(est_sources[:, :, 1].detach().cpu().squeeze(), rate=8000)
```

```python
print(est_sources.shape)
```

```python
model_2 = separator.from_hparams(source="speechbrain/sepformer-wham16k-enhancement", savedir='pretrained_models/sepformer-wham16k-enhancement')

# for custom file, change path
est_sources_2 = model_2.separate_file(path='speechbrain/sepformer-wham16k-enhancement/example_wham16k.wav') 
```

```python
print(est_sources_2.shape)

signal_2 = read_audio("example_wham16k.wav").squeeze()
IPython.display.display(Audio(signal_2, rate=16000))

IPython.display.display(Audio(est_sources_2[:, :, 0].detach().cpu().squeeze(), rate=16000))
```

```python
model_3 = separator.from_hparams(source="speechbrain/sepformer-whamr16k", savedir='pretrained_models/sepformer-whamr16k')

# for custom file, change path
est_sources_3 = model_3.separate_file(path='speechbrain/sepformer-whamr16k/test_mixture16k.wav') 
```

```python
print(est_sources_3.shape)

signal_3 = read_audio("test_mixture16k.wav").squeeze()
IPython.display.display(Audio(signal_3, rate=16000))

IPython.display.display(Audio(est_sources_3[:, :, 0].detach().cpu().squeeze(), rate=16000))
IPython.display.display(Audio(est_sources_3[:, :, 1].detach().cpu().squeeze(), rate=16000))
```

```python
model_4 = separator.from_hparams(source="speechbrain/sepformer-libri3mix", savedir='pretrained_models/sepformer-libri3mix')

est_sources_4 = model_4.separate_file(path='speechbrain/sepformer-wsj03mix/test_mixture_3spks.wav', savedir='pretrained_models/sepformer-libri3mix') 
```

```python
print(est_sources_4.shape)

signal_4 = read_audio("pretrained_models/sepformer-libri3mix/test_mixture_3spks.wav").squeeze()
IPython.display.display(Audio(signal_4, rate=8000))

IPython.display.display(Audio(est_sources_4[:, :, 0].detach().cpu().squeeze(), rate=8000))
IPython.display.display(Audio(est_sources_4[:, :, 1].detach().cpu().squeeze(), rate=8000))
IPython.display.display(Audio(est_sources_4[:, :, 2].detach().cpu().squeeze(), rate=8000))
```
